<h2>Skewness</h2>
Skewness is the measure of the symmetry of a distribution compared to standard normal distribution<br>    
+ive - right skewed (mean is to the right of mode/median). Long tail in the +ive direction.<br>
0 - symmetric<br>
-ive - left skewed (mean is to the left of mode/median). Long tail in the -ive direction.

```python
df.skew()
```

<h2>Kurtosis</h2>
Kurtosis is a measure of the flatness or peakedness of a distribution compared to the normal distribution.<br>
+ive - Leptokurtosis (sharper/spikier peak compared to the normal dist.)
0 - Mesokurtic (normal dist.)<br>
-ive - Platykurtic (flatter peak compared to the normal dist.) eg. Uniform distribution<br>

```python
df.kurtosis()
```

<h2>Line Plot</h2>

```python
# Plotting with index along the x-axis
df['installment'].plot(figsize=(12, 5), color='black') # color and figsize changed

plt.xlim(0, 1000) # range for x-axis
plt.ylim(0, 1000) # range for x-axis
plt.xlabel('index')
plt.ylabel('Installment'); # ";" prevents object info from displaying
```

<h2>Scatter Plot</h2>

```python
# plotting one variable against the other
df.plot.scatter('fico', 'installment', figsize=(8, 5))

# The x and y labels are automatically taken from the column names
```

<h2>Box Plot</h2>

```python
# Box plot of a column
df['installment'].plot.box(figsize=(8, 5));
```

```python
# Box plot of all the columns with numerical data
df.boxplot(figsize=(16, 5)) # or df.plot.box()
```

<h2>Histogram</h2>

```python
df['installment'].hist(bins=30, figsize=(8, 5)); # we can specify the number of bins
```

```python
ax = df['installment'].hist(bins=30, grid=False, color='green', figsize=(8, 5)) # grid turned off and color changed

ax.set_xlabel('installment')
ax.set_ylabel('fico')

ax.set_xlim(0, 1000) # limiting display range to 0-70 for the x-axis
ax.set_ylim(0, 1000); # limiting display range to 0-120 for the y-axis
```

<h2>Bar Plot</h2>

```python
# To stack values from multiple columns
ax = df[['installment', 'fico']][:5].plot.bar(stacked=True)

ax.set_xlabel("Index")
ax.set_ylabel("Count");
```

<h2>Multiple Plots</h2>

```python

fig, axes = plt.subplots(2, 2, figsize=(12, 8))
# or fig, (ax1, ax2, ax3, ax4) = plt.subplots(2, 2, figsize=(12, 8))

# axes is the axes object(s). It can be a single object or an array of objects. 
# In this case, it is an array of dimension 2-by-2

df['installment'].plot(ax = axes[0][0], style='.', color='red') # top left
df['fico'].plot(ax = axes[0][1], style='.', color='blue') # top right

df['int.rate'].plot.hist(bins=30, ax = axes[1][0], color='black') # bottom left
df['dti'].plot.hist(bins=20, ax = axes[1][1], color='gray') # bottom right

axes[0][0].set_xlabel('index')
axes[0][1].set_xlabel('index')

axes[0][0].set_ylabel('installment')
axes[0][1].set_ylabel('fico')

axes[1][0].set_xlabel('int.rate')
axes[1][1].set_xlabel('dti')

axes[0][0].set_ylim(0, 1000)
axes[0][1].set_ylim(600, 800)

axes[1][0].set_xlim(0.05, 0.2)
axes[1][1].set_xlim(0, 30)

fig.tight_layout()
```
